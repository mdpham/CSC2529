```python
import numpy as np
from numpy.fft import fft2, ifft2, fftshift, ifftshift
import skimage.io as io
from skimage.util import random_noise
import matplotlib.pyplot as plt
import skimage.metrics as metrics
# from skimage.metrics import peak_signal_noise_ratio as compute_psnr
from pypher.pypher import psf2otf

from pdb import set_trace
from fspecial import fspecial_gaussian_2d
```

```python
img = io.imread('birds_gray.png').astype(float)/255
io.imshow(img)
```

```python
def make_filter(sigma=5, filtSize=35):
    filt = fspecial_gaussian_2d((filtSize, filtSize), sigma)    
    print(f'filter sigma {sigma}, filter size {filtSize}')
    return filt

def blur_filter(img, sigma=5):
    psf = make_filter()
    otf = psf2otf(psf, shape=np.shape(img))
    img_fourier = fft2(img)
    img_lowpass = ifft2(np.multiply(img_fourier, otf))
    return img_lowpass.real

def add_noise(img, sigma=0):
    noise = np.zeros_like(img) + sigma * np.random.randn(*img.shape)
    return noise + img
```

```python
io.imshow(blur_filter(img))
```

```python
io.imshow(add_noise(blur_filter(img), 0.01), cmap=plt.cm.gray)
```

```python
def inverse_filter_simple(img, blur_kernel):
    img_fourier = fft2(img)
    deblur_fourier = np.divide(img_fourier, blur_kernel)
    deblur = ifft2(deblur_fourier)
    return deblur.real

def inverse_filter_wiener(img, blur_kernel, sigma_noise):
    img_fourier = fft2(img)
    snr = np.mean(img)/sigma_noise
    h_2 = np.square(np.absolute(blur_kernel))
    wiener_kernel = np.divide(h_2, np.multiply(blur_kernel, np.add(h_2, 1./snr)))
    deblur_fourier = np.multiply(img_fourier, wiener_kernel)
    deblur = ifft2(deblur_fourier)
    return deblur.real
```

```python
sigma_noise = 0.001
sigma_blur = 5
blur_kernel = psf2otf(make_filter(sigma=sigma_blur), shape=np.shape(img))
unfilt = add_noise(blur_filter(img), sigma=sigma_noise)
filt_simple = inverse_filter_simple(unfilt, blur_kernel)
filt_wiener = inverse_filter_wiener(unfilt, blur_kernel, sigma_noise)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
ax[0].imshow(filt_simple, cmap=plt.cm.gray)
ax[1].imshow(filt_wiener, cmap=plt.cm.gray)

fig.tight_layout()
plt.show()
```

```python
blur = blur_filter(img)
blur_kernel = psf2otf(make_filter(), shape=np.shape(img))

results = {}

sigmas_noise = [0, 0.001, 0.01, 0.1]
for sigma_noise in sigmas_noise:
    unfilt = add_noise(blur, sigma_noise)
    io.imsave(f't2/{sigma_noise}_noise.png', unfilt)
    filt_simple = inverse_filter_simple(unfilt, blur_kernel)
    filt_wiener = inverse_filter_wiener(unfilt, blur_kernel, sigma_noise)
    results[f'{sigma_noise}_simple'] = filt_simple
    results[f'{sigma_noise}_wiener'] = filt_wiener
    
for method in ['simple', 'wiener']:
    for sigma_noise in sigmas_noise:
        key = f'{sigma_noise}_{method}'
        filt = results[key]
        mse = metrics.mean_squared_error(img, filt)
        psnr = metrics.peak_signal_noise_ratio(img, filt)
        print(key, psnr)
```

```python
keys = list(results.keys())
fig, axes = plt.subplots(4, 2, figsize=(8, 8))
ax = axes.ravel()
for i in range(len(keys)):
    key = keys[i]
    filt = results[key]
    ax[i].imshow(filt, cmap=plt.cm.gray)
    ax[i].set_title(key)
    
    io.imsave(f't2/{key}.png', filt)

fig.tight_layout()
plt.show()
```
