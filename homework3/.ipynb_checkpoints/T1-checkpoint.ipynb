```python
import numpy as np
from numpy.fft import fft2, ifft2, fftshift, ifftshift
import skimage.io as io
from scipy.signal import convolve2d
from time import process_time
from pypher.pypher import psf2otf
import matplotlib.pyplot as plt

from fspecial import fspecial_gaussian_2d
import time
```

```python
img = io.imread('birds_gray.png').astype(float)/255
io.imshow(img)
```

```python
def make_filter(sigma, filtSize=None):
    if (filtSize is None):
        filtSize = np.ceil(9 * sigma).astype(int)
    filt = fspecial_gaussian_2d((filtSize, filtSize), sigma)    
    print(f'filter sigma {sigma}, filter size {filtSize}')
    return filt

def lowpass_filter_primal(img, sigma, filtSize=None):
    lp_filt = make_filter(sigma, filtSize)
    img_lowpass = convolve2d(img, lp_filt, mode='same')    
    return img_lowpass

def highpass_filter_primal(img, sigma, filtSize=None):
    img_highpass = np.subtract(img, lowpass_filter_primal(img, sigma, filtSize))
    return img_highpass
```

```python
def lowpass_filter_fourier(img, sigma):
    psf = make_filter(sigma)
    otf = psf2otf(psf, shape=np.shape(img))
    img_fourier = fft2(img)
    img_lowpass = ifft2(np.multiply(img_fourier, otf))
    return img_lowpass.real

def highpass_filter_fourier(img, sigma, filtSize=None):
    psf = make_filter(sigma, filtSize)
    otf = psf2otf(psf, shape=np.shape(img))
    img_fourier = fft2(img)
    img_highpass = ifft2(np.multiply(img_fourier, np.subtract(1, otf)))
    return img_highpass.real
```

```python
lp_results = {}
hp_results = {}
sigmas = [0.1, 1, 10, 15]
for sigma in sigmas:
    # low pass primal
    start = time.perf_counter()
    img_lp_p = lowpass_filter_primal(img, sigma)
    end = time.perf_counter()
    lp_results[f'{sigma}_primal'] = {'img': img_lp_p, 'time': end-start}
    # low pass fourier
    start = time.perf_counter()
    img_lp_f = lowpass_filter_fourier(img, sigma)
    end = time.perf_counter()
    lp_results[f'{sigma}_fourier'] = {'img': img_lp_p, 'time': end-start}
    # high pass primal
    start = time.perf_counter()
    img_hp_p = highpass_filter_primal(img, sigma, filtSize=101)
    end = time.perf_counter()
    hp_results[f'{sigma}_primal'] = {'img': img_hp_p, 'time': end-start}
    # high pass fourier
    start = time.perf_counter()
    img_hp_f = highpass_filter_fourier(img, sigma, filtSize=101)
    end = time.perf_counter()
    hp_results[f'{sigma}_fourier'] = {'img': img_hp_f, 'time': end-start}    
```

```python
domains = ['primal', 'fourier']
for domain in domains:
    for sigma in sigmas:
        key = f'{sigma}_{domain}'
        print(key, np.log10(1+hp_results[key]['time']))
```
