```python
import numpy as np
from tqdm import tqdm
from time import time
import matplotlib.pyplot as plt
from functools import partial
```

```python
def grad_l2(A, x, b):
#     TODO: return the gradient of 0.5 * ||Ax - b||_2^2
    aTax = np.matmul(np.matmul(A.T, A), x)
    aTb = np.matmul(A.T, b)
    return np.subtract(aTax, aTb)

def residual_l2(A, x, b):
    return 0.5 * np.linalg.norm(A @ x - b)**2


def run_gd(A, b, step_size=1e-4, num_iters=1500, grad_fn=grad_l2, residual=residual_l2):
    ''' Run gradient descent to solve Ax = b

    Parameters
    ----------
    A : matrix of size (N_measurements, N_dim)
    b : observations of (N_measurements, 1)
    step_size : gradient descent step size
    num_iters : number of iterations of gradient descent
    grad_fn : function to compute the gradient
    residual : function to compute the residual

    Returns
    -------
    x
        output matrix of size (N_dim)

    residual
        list of calculated residuals at each iteration

    timing
        time to execute each iteration (should be cumulative to each iteration)

    '''
    
    results = []

    # initialize x to zero
    x = np.zeros((A.shape[1], 1))
    
    for i in range(num_iters):
        # TODO: complete the gradient descent algorithm here
        # you can also complete and use the grad_l2 and residual_l2 functions

        # this function can return a list of residuals and timings at each iteration so
        # you can plot them and include them in your report

        # don't forget to also implement the stochastic gradient descent version of this function!
        time_start = time()
        x_new = x - step_size*grad_fn(A, x, b)
        time_end = time()
        x = x_new
        
        resid = residual_l2(A, x, b)
        time_elapsed = time_end - time_start
        results.append({'time': time_elapsed, 'res': resid})

    return results
    
def run_sgd(A, b, batch_size=10, step_size=1e-4, num_iters=1500, grad_fn=grad_l2, residual=residual_l2):
    print(f'sgd batch_size {batch_size}')
    results = []
    x = np.zeros((A.shape[1], 1))
    def sample_indices(A, batch_size=batch_size):
        batch_idx = np.random.choice(range(A.shape[0]), size=batch_size, replace=False)
        A_sample = A[batch_idx, :]
        return batch_idx, A_sample

    for i in range(num_iters):
        batch_idx, A_sample = sample_indices(A, batch_size)
        b_sample = b[batch_idx]
        time_start = time()
        x_new = x - step_size*grad_fn(A_sample, x, b_sample)
        time_end = time()
        x = x_new
        
        resid = residual_l2(A, x, b)
        time_elapsed = time_end - time_start
        results.append({'time': time_elapsed, 'res': resid})
    return results


def run_lsq(A, b):
    ''' Numpy's implementation of least squares which uses SVD & matrix factorization from LAPACK '''
    x, resid, _, _ = np.linalg.lstsq(A, b, rcond=None)
    return x, 0.5 * resid

```

```python
batch_sizes = [10, 100, 1000]
result_sgd = {}
for batch_size in batch_sizes:
    result_sgd[f'{batch_size}'] = run_sgd(A, b, batch_size = batch_size)
```

```python
def plot_residuals(result_gd, result_lsq, result_sgd):
    #     Get x-axis for iteration and times
    x_times = [0]
    for i in range(1, len(results_gd)):
        x_times.append(x_times[i-1] + results_gd[i]['time'])
    x_iterations = list(range(len(results_gd)))
    
    
    #     Set up plots and axis scales
    fig, axes = plt.subplots(2, 1, figsize=(8, 8))
    ax = axes.ravel()
    axs_iter = ax[0]
    axs_iter.set_ylabel('residual')
    axs_iter.set_xlabel('iterations')
    axs_iter.set_yscale('log')
    
    axs_time = ax[1]
    axs_time.set_ylabel('residual')
    axs_time.set_xlabel('time (s)')
    axs_time.set_yscale('log')
    axs_time.set_xscale('log')
    
    #     Get GD results
    y_residuals_gd = [r['res'] for r in results_gd]
    #     Get SVD results
    y_residuals_svd = np.tile(result_lsq['res'], len(results_gd))
    
    #     Plot SVD
    axs_iter.plot(x_iterations, y_residuals_svd, color='blue')
    #     Plot GD
    axs_time.plot(x_times, y_residuals_gd, color='orange')
    axs_iter.plot(x_iterations, y_residuals_gd, color='orange')
    #     Plot SGD
    sgd_color = {'10': 'green', '100': 'red', '1000': 'purple'}
    for batch_size in batch_sizes:
        #     Get SGD results
        y_residuals_sgd = [r['res'] for r in result_sgd[f'{batch_size}']]
        axs_time.plot(x_times, y_residuals_sgd, color=sgd_color[f'{batch_size}'])
        axs_iter.plot(x_iterations, y_residuals_sgd, color=sgd_color[f'{batch_size}'])
    
    fig.tight_layout()
    plt.show()
    
    pass

plot_residuals(result_gd, result_lsq, result_sgd)
```
